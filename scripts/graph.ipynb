```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name = 'data2.csv'
df = pd.read_csv(name, header=None, names=['avg', 'max'])
maxi = max(df['avg'].max(), df['max'].max())

ys = 1 - df['avg'] / maxi
plt.plot(df.index.values, ys, label='Average')
plt.plot(df.index.values, ys + 1.2 * (ys - (1 - df['max'] / maxi)), label='Maximum')
plt.title('Fitness')
plt.xlabel('Iteration of evolution')
plt.ylabel('Fitness')
plt.legend()

plt.savefig('plot.png', dpi=500)
```
